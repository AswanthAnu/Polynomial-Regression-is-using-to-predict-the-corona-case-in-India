### Handling missing values

```python
# filling missing values with zeros.
df = df.fillna(0)
```

```python
df.isnull().sum()
```

### Splitting the dataset

```python
X = df.drop(['total_cases'], axis=1)
y = df['total_cases']
```

```python
### train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=0)
```

### Feature Engineering

```python
# Feature selection using ANOVA

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
selector = SelectKBest(f_classif, k=2)

selector.fit_transform(X_train, y_train)
```

```python
X_train_fs = selector.transform(X_train)
X_test_fs = selector.transform(X_test)
```

```python
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X_train_fs)

# transform training data
X_train_norm = norm.transform(X_train_fs)

# transform testing dataabs
X_test_norm = norm.transform(X_test_fs)

```

### Modeling

```python
# Create model using polynomial regression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree =3) #3rd degree polynomial
X_train_norm = poly.fit_transform(X_train_norm)

lin_reg = LinearRegression()
lin_reg.fit(X_train_norm, y_train)
lin_reg.score(X_train_norm, y_train)
```
